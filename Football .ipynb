```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Footballer.csv")
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
df.drop(columns=["Unnamed: 0", "Photo", "Flag	"], inplace=True)
```

```python
df.drop(columns=["ID"], inplace=True)
```

```python
# numerical_columns = df.select_dtypes.include(["int", "float"])
numeric_columns = df.select_dtypes(include=['number']).columns
numeric_columns
```

```python
# Generate 42 different random colors
random_colors = np.random.rand(42, 3)  # Generating RGB values
```

```python
for i, column in enumerate(numeric_columns):
    plt.figure(figsize=(10, 6))  # Set figure size
    sns.boxplot(x=column, data=df, color=random_colors[i % len(random_colors)])  # Apply random color
    plt.title(column, fontsize=16)  # Set title with column name
    plt.xlabel('Values', fontsize=14)  # Set x-axis label
    plt.ylabel('Frequency', fontsize=14)  # Set y-axis label
    plt.xticks(fontsize=12)  # Set x-axis tick labels font size
    plt.yticks(fontsize=12)  # Set y-axis tick labels font size
    plt.show()
```

```python
import itertools
import warnings
warnings.filterwarnings("ignore")



# Get all combinations of column pairs
column_pairs = list(itertools.combinations(df.columns, 2))

# Define the number of rows and columns for subplots
num_rows = 10  # Number of rows
num_cols = 7   # Number of columns

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 25))

# Flatten axes if necessary
axes = axes.flatten()

# Plot pairwise bivariate analyses for each combination
for i, (col1, col2) in enumerate(column_pairs):
    ax = axes[i]
    sns.scatterplot(x=col1, y=col2, data=df, ax=ax)
    ax.set_title(f'{col1} vs {col2}')
    ax.set_xlabel(col1)
    ax.set_ylabel(col2)

# Remove empty subplots
for j in range(len(column_pairs), num_rows * num_cols):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
