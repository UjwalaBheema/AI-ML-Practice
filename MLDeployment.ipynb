```python
# !pip install PyPDF2



import PyPDF2
# from PyPDF2 import PdfReader
# from PyPDF2 import PdfMerger, PdfReader



# def select_pages(pdf_path, start_page, end_page):
#     selected_pages = []
#     with open(pdf_path, 'rb') as file:
#         reader = PyPDF2.PdfReader(file)
#         for page_num in range(start_page - 1, end_page):
#             page = reader.pages[page_num]
#             selected_pages.append(page)
#     return selected_pages

# def merge_pdfs(pdf_list, output_path):
#     merger = PyPDF2.PdfMerger()
#     for pdf_path in pdf_list:
#         merger.merge(pdf_path)
#     with open(output_path, 'wb') as output_file:
#         merger.write(output_file)

# # Example usage:
# pdf1_pages = select_pages('', 1, 3)  # Select pages 1 to 3 from file1.pdf
# # pdf2_pages = select_pages('file2.pdf', 2, 4)  # Select pages 2 to 4 from file2.pdf

# # Merge selected pages
# merge_pdfs(pdf1_pages, 'merged_output.pdf')


import PyPDF2

def select_pages(pdf_path, start_page, end_page):
    selected_pages = []
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        if start_page < 1 or end_page > num_pages:
            raise ValueError("Invalid page range.")
        for page_num in range(start_page - 1, end_page):
            page = reader.pages[page_num]
            selected_pages.append(page)
    return selected_pages

def merge_pdfs(pdf_list, output_path):
    writer = PyPDF2.PdfWriter()
    for pdf_pages in pdf_list:
        for page in pdf_pages:
            writer.add_page(page)
    with open(output_path, 'wb') as output_file:
        writer.write(output_file)



# Example usage:
try:
    pdf1_pages = select_pages('/Users/bheema.ujwala/Downloads/Jan 25th Confirmation for Booking ID.pdf', 1, 3)  # Select pages 1 to 3 from file1.pdf
#     pdf2_pages = select_pages('file2.pdf', 2, 4)  # Select pages 2 to 4 from file2.pdf

    # Merge selected pages
    merge_pdfs([pdf1_pages], 'merged_output.pdf')

except ValueError as e:
    print("Error:", e)

    

# def select_pages(pdf_path, start_page, end_page):
#     selected_pages = []
#     with open(pdf_path, 'rb') as file:
#         reader = PyPDF2.PdfReader(file)
#         num_pages = len(reader.pages)
#         if start_page < 1 or end_page > num_pages:
#             raise ValueError("Invalid page range.")
#         for page_num in range(start_page - 1, end_page):
#             page = reader.pages[page_num]
#             selected_pages.append(page)
#     return selected_pages

# def merge_pdfs(pdf_list, output_path):
#     writer = PyPDF2.PdfWriter()
#     for pdf_path in pdf_list:
#         with open(pdf_path, 'rb') as file:
#             reader = PyPDF2.PdfReader(file)
#             for page in reader.pages:
#                 writer.add_page(page)
#     with open(output_path, 'wb') as output_file:
#         writer.write(output_file)

# # Example usage:
# try:
#     pdf1_pages = select_pages('/Users/bheema.ujwala/Downloads/Jan 25th Confirmation for Booking ID.pdf', 1, 3)  # Select pages 1 to 3 from file1.pdf
# #     pdf2_pages = select_pages('file2.pdf', 2, 4)  # Select pages 2 to 4 from file2.pdf

#     # Merge selected pages
#     merge_pdfs(pdf1_pages, 'merged_output.pdf')

# except ValueError as e:
#     print("Error:", e)


```

```python
df = pd.read_csv("titanic-training-data.csv")
```

```python
df = df.drop('Cabin', axis=1)
```

```python
df = df.dropna()
df.shape
```

```python
X=df[["Pclass", "Sex","Age", "SibSp", "Parch", "Fare"]]
y=df["Survived"]
```

```python
df.dtypes
```

```python
label_encoder = LabelEncoder()
```

```python
X["Sex"]=label_encoder.fit_transform(X["Sex"])
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```
