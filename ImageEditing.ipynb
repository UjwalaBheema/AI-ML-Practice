```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('0-04-08-0a3ce9e823a38581b790b5658a049e8d4d0c2c29228b320079e3efcd261a7d3a_1c6db8f791a6ef.jpg')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap="gray")
plt.axis('off')
plt.show()
```

```python
res=cv2.resize(image, (300,200))
roi=image[0:150,100:200]

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Resized Image")


plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Cropped Image")
plt.show()
```

```python
res=cv2.resize(image, (5000,500))
roi=image[100:150,20:100]

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Resized Image")


plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Cropped Image")
plt.show()
```

```python
rows, cols, _ = image.shape

rows, cols, _
```

```python
rotation_matrix=cv2.getRotationMatrix2D((cols/2, rows/2),45,1)
rotated_image = cv2.warpAffine(image,rotation_matrix,(cols,rows))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
rotation_matrix=cv2.getRotationMatrix2D((cols/2, rows/2),180,1)
rotated_image = cv2.warpAffine(image,rotation_matrix,(cols,rows))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
rotation_matrix=cv2.getRotationMatrix2D((cols/2, rows/2),90,2) # changed scale to 2 to zoom
rotated_image = cv2.warpAffine(image,rotation_matrix,(cols,rows))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
rotation_matrix=cv2.getRotationMatrix2D((cols/2, rows/2),180,100) # More scale
rotated_image = cv2.warpAffine(image,rotation_matrix,(cols,rows))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
rotation_matrix=cv2.getRotationMatrix2D((cols/2, rows/2),180,100) # More scale
rotated_image = cv2.warpAffine(image,rotation_matrix,(cols,rows))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(image,(15,15),0)


plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")


plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Blurred Image")
plt.show()
```

```python
alpha = 2.5  # Contrast control 
beta = 40   # Brightness control

adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Contrast Image")
plt.show()
```

```python
cartoon_image = cv2.stylization(image, sigma_s=150, sigma_r=0.25)
plt.imshow(cv2.cvtColor(cartoon_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
