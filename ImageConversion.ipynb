```python
import cv2
import numpy as np
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

# Function to apply grayscale filter
def apply_grayscale(img):
    print("apply_grayscaleapply_grayscale")
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Function to apply sepia filter
def apply_sepia(img):
    print("apply_sepiaapply_sepiaapply_sepia")
    sepia_filter = np.array([[0.272, 0.534, 0.131],
                             [0.349, 0.686, 0.168],
                             [0.393, 0.769, 0.189]])
    sepia_image = cv2.transform(img, sepia_filter)
    sepia_image[np.where(sepia_image > 255)] = 255  # Cap values to 255
    return sepia_image.astype(np.uint8)
```

```python
# Function to open an image file
def open_image():
    global current_image
    file_path = filedialog.askopenfilename()
    if file_path:
        current_image = cv2.imread(file_path)
        update_display(current_image)


# Function to update the displayed image
def update_display(image):
    pil_image = Image.fromarray(image)
    pil_image = pil_image.resize((300, 300))
    tk_image = ImageTk.PhotoImage(pil_image)
    image_label.config(image=tk_image)
    image_label.image = tk_image

# Create the Tkinter root window
root = tk.Tk()
root.title("Image Filters")

# Create buttons and labels
open_button = tk.Button(root, text="Open Image", command=open_image)
open_button.pack()
```

```python
image_label = tk.Label(root)
image_label.pack()

# # # Create the Tkinter root window
# root = tk.Tk()
# root.title("Image Sketch Converter")

# root = tk.Tk()
# root.title("Image Sketch Converter")

# # Create buttons and labels
# open_button = tk.Button(root, text="Open Image", command=open_image)
# open_button.pack()

# original_image_label = tk.Label(root)
# original_image_label.pack(side=tk.LEFT)

# sketch_image_label = tk.Label(root)
# sketch_image_label.pack(side=tk.RIGHT)



# # # Create buttons and labels
# # open_button = tk.Button(root, text="Open Image", command=open_image)
# # open_button.pack()

# # original_image_label = tk.Label(root)
# # original_image_label.pack(side=tk.LEFT)

# # sketch_image_label = tk.Label(root)
# # sketch_image_label.pack(side=tk.RIGHT)

# # # # Create buttons for each filter
# # grayscale_button = tk.Button(root, text="Grayscale", command=lambda: apply_filter(apply_grayscale))
# # grayscale_button.pack(side=tk.LEFT)

# # sepia_button = tk.Button(root, text="Sepia", command=lambda: apply_filter(apply_sepia))
# # sepia_button.pack(side=tk.LEFT)

# # Create buttons for each filter
# filters = [("Grayscale", apply_grayscale),
#            ("Sepia", apply_sepia)]

# for filter_name, filter_func in filters:
# #     tk.Button(root, text="Grayscale", command=lambda: apply_filter(apply_grayscale))
#     print("filter_name,filter_func", filter_name, filter_func)
#     button = tk.Button(root, text=filter_name, command=lambda: apply_filter(filter_func))
#     button.pack()

# # Function to apply a selected filter
def apply_filter(filter_func):
    global current_image

    if current_image is None:
        return

    filtered_img = filter_func(current_image)
    update_display(filtered_img)

    
# # Start the Tkinter main loop
# current_image = None
# root.mainloop()


root = tk.Tk()
root.title("Image Filters")

# Button to open image file
open_button = tk.Button(root, text="Open Image", command=open_image)
open_button.pack()

# Button to apply filters
filter_buttons = [
    ("Grayscale", "grayscale"),
    ("Negative", "negative"),
    ("Blur", "blur"),
    ("Edges", "edges")
]

for text, filter_type in filter_buttons:
    button = tk.Button(root, text=text, command=lambda filter_type=filter_type: apply_filter(filter_type))
    button.pack()

# Label to display the image
img_label = tk.Label(root)
img_label.pack()

# Original image
current_image = None

# Start the GUI event loop
root.mainloop()
```
